#### Linear Functions and Arithmetic Sequences

In general, we recognize sequences with constant change as "linear" functions.  We can define these by thinking recursively and using the representation $$f(n) = \text{starting value} + \text{change} \times n$$

For example, the sequence $a_n = [3, 5, 7, 9, 11, 13]$, we have:

$$\text{starting value} = 3$$
$$\text{change} = 2$$

thus;

$$f(n) = 3 + 2n$$

**Problem**: Determine a linear function to model the data.  

One approach would be to use our observations about linear functions:

$$f(n) = \text{starting value} + \text{change} \times n$$

The closing price on the first day of the data is 370.869995, and on the twentieth day is 408.589996.  

- This represents a change of (408.589996 - 370.869995)/19 = 2.0447371633429277

thus;

$$f(n) = 370.869995 + 2.0447371633429277n$$

````python
#difference from last day to first day divided by number of days
(msft_history['Close'][-1] - msft_history['Close'][0])/19
````

````python
#define the function
def f(n):
    return 370.869995+2.0447371633429277*n
````

````python
np.arange(19)
````

````python
n = np.arange(19)
````

````python
f(n)
````

````python
plt.plot(msft_history['Close'].values, '-o', label = 'data')
plt.title('Microsoft Closing Prices')
plt.grid()
plt.xticks(rotation = 90)
plt.plot(n, f(n), label = 'model')
plt.legend()
plt.xticks(n);
````

````python
#approximate value from model for day index 10
f(10)
````

````python
#true value on 10th index
msft_history['Close'][10]
````

````python
predicted = f(10)
actual = msft_history['Close'][10]
#error
predicted - actual
````

#### Slope and Intercept

The change of a linear function is often talked about as the *slope* of the line.  In general, given any two points on a line -- $(x_1, y_1)$ and $(x_2, y_2)$ -- , the slope is the average difference between those points:

$$\text{slope} = \frac{y_2 - y_1}{x_2 - x_1}$$

and by rearranging this equation you can write an equation for the line as:

$$y = \text{slope} \times (x - x_1) + y_1$$

For example, suppose you want an equation between the points $(3, 5)$ and $(8, 12)$.  

First we determine the slope $\frac{12 - 5}{8 - 3} = \frac{7}{5}$, then plug into our equation $$y = \frac{7}{5}(x - 3) + 5$$

#### Geometric Sequences and Exponential Functions

Consider the sequence formed by $a_0 = 10$ and $a_{n + 1} = a_n \times 1.1$

$$10.0, 11.0, 12.1, 13.31, 14.641, 16.105, 17.716, 19.487, 21.436, 23.579, ...$$

What would $a_{20}$ be equal to?  $a_{100}$?  If you recognize the repeated multiplication happening you can rewrite the terms of the sequence as

$$f(n) = 10(1.1^n)$$

Define and plot the function for $n \in [0, 20]$.

````python
#define the function
def f(x): return 10*1.1**x
````

````python
#domain
x = np.linspace(0, 20, 100)
````

````python
#plot
plt.plot(x, f(x))
````
